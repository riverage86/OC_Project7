```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime 
from datetime import date
import math
import pandas_datareader as web
import yfinance as yf
```

```python
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
start_date = datetime.datetime(2010, 1, 2)
end_date = datetime.datetime.today()
```

```python
raw_data = yf.download('^GSPC', start=start_date, end=end_date)
raw_data.to_parquet('sp500_data.parquet')
```

```python
raw_data.head()
```

```python
features = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Adj Close', 'Volume']
```

```python
data = pd.read_parquet('sp500_data.parquet')
```

```python
data.head()
```

```python
print('There are {} number of days in the dataset.'.format(data.shape[0]))
```

```python
plt.figure(figsize=(28, 12))#, dpi=100)
plt.plot(data.index, data['Adj Close'], label='SP500')
plt.xlabel('Date')
plt.ylabel('Rs')
plt.title('sp500')
plt.legend()
plt.show()
```

```python
df = data[['Adj Close', 'Open', 'High', 'Low', 'Close', 'Volume']]
```

```python
df.head()
```

```python
df = df.dropna()
df.head()
```

```python
print('Total dataset has {} samples, and {} features.'.format(df.shape[0], \
                                                              df.shape[1]))
```

```python
data_training = df[df.index < '2019-01-31'].copy()
data_training.head()
```

```python
data_testing = df[df.index >= '2019-01-31'].copy()
data_testing.head()
```

```python
data_training.shape, data_testing.shape
```

```python
scalar = MinMaxScaler()

data_training_scaled = scalar.fit_transform(data_training)
print(data_training_scaled.shape)
data_training_scaled
```

```python
# window size is 60, nb sample = 1935 (data_training.shape[0]), y_train is price (first colonne, 0)
```

```python
X_train = []
y_train = []
for i in range(60, data_training.shape[0]):
    X_train.append(data_training_scaled[i-60: i])
    y_train.append(data_training_scaled[i, 0])
```

```python
X_train, y_train = np.array(X_train), np.array(y_train)
X_train.shape, y_train.shape
```

```python
# add the winodw before test data
dt = pd.concat([data_training.tail(60), data_testing], ignore_index=True)
dt.head()
inputs = scalar.fit_transform(dt)

X_test = []
y_test = []

for i in range(60, inputs.shape[0]):
    X_test.append(inputs[i-60:i])
    y_test.append(inputs[i, 0])
    
X_test, y_test = np.array(X_test), np.array(y_test)
X_test.shape, y_test.shape
```

```python
regressor = Sequential()

regressor.add(LSTM(units = 50, activation = 'relu', input_shape = (X_train.shape[1], X_train.shape[2])))
regressor.add(Dropout(0.5))

#regressor.add(LSTM(units = 60, activation = 'relu', return_sequences = True))
#regressor.add(Dropout(0.3))

#regressor.add(LSTM(units = 80, activation = 'relu', return_sequences = True))
#regressor.add(Dropout(0.4))

#regressor.add(LSTM(units = 120, activation = 'relu'))
#regressor.add(Dropout(0.5))

regressor.add(Dense(units = 1))
```

```python
regressor.summary()
```

```python
# Compiling the RNN
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
history = regressor.fit(X_train, y_train, epochs=50, batch_size = 64,shuffle=False, validation_data=(X_test, y_test),verbose=1) 
```

```python
# Plotting training and validation loss over epochs
import matplotlib.pyplot as plt

plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
#plt.yscale('log')  # Set y-axis to logarithmic scale
plt.legend()
plt.show()
```

```python
def lstm_mse_rsme(lstm_training): 
    fig, ax = plt.subplots(figsize=(12, 4))

    # transforme MSE en RMSE
    loss_history = pd.DataFrame(lstm_training.history).pow(.5)
    loss_history.index += 1

#calcule la valeur minimale de la colonne val_loss dans le DataFrame loss_history, 
#ce qui correspond au meilleur RMSE de validation atteint par le modle.
    best_rmse = loss_history.val_loss.min()
    best_epoch = loss_history.val_loss.idxmin()

    title = f'5-Epoch Rolling RMSE (Best Validation RMSE: {best_rmse:.4%})'
    loss_history.columns=['Training RMSE', 'Validation RMSE']
    loss_history.rolling(5).mean().plot(logy=True, lw=2, title=title, ax=ax)

#une ligne verticale pointille (ls='--') sur le graphique pour marquer
#l'poque avec le meilleur RMSE de validation (best_epoch).
    ax.axvline(best_epoch, ls='--', lw=1, c='k')

    sns.despine()
    fig.tight_layout()
    return best_rmse
```

```python
rmse = lstm_mse_rsme(history)
```

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred[:3]
```

```python
scale = 1/scalar.scale_[0]
```

```python
y_pred = y_pred*scale
y_test = y_test*scale
```
